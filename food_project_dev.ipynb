```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn
```

```python
def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement
    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```

```python
def main():
    database = "sqlite/db/pythonsqlite.db"

    sql_create_food_prices_table = """CREATE TABLE IF NOT EXISTS food_prices (
                                    id integer PRIMARY KEY,
                                    cm_name text NOT NULL,
                                    adm0_name text NOT NULL,
                                    um_name text NOT NULL,
                                    mp_year text NOT NULL,
                                    mp_price integer NOT NULL
                                );"""

    # create a database connection
    conn = create_connection(database)

    # create tables
    if conn is not None:
        # create food_prices table
        create_table(conn, sql_create_food_prices_table)

    else:
        print("Error! cannot create the database connection.")
```

```python
if __name__ == '__main__':
    main()
```
